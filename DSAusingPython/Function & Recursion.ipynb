```python
# Check Palindrome

def helper(str):
    size=len(str)
    for i in range(size//2):
        if(str[i]!=str[size-1-i]):
            return False
    return True

def helper_rec(str, i, j): #by recursion
    
    if(i==j):
        return True
    
    if(str[i]!=str[j]):
        return False
    else:
        return helper_rec(str, i+1, j-1)

str=input("Enter any String:\n")
print(helper(str))
print(helper_rec(str, 0, len(str)-1))
```

```python
# Factorial

def helper(n): #non-recursive
    if(n<=0):
        return 0;
    else:
        ans=1
        for i in range(1, n+1):
            ans*=i
        return ans
    
def helper_rec(n): #recursive
    if(n<=1):
        return 1;
    else:
        return n*helper_rec(n-1);

n=int(input("Enter any number:\n"))
print(helper(n))
print(helper_rec(n))
```

```python
# add 1 - N

def helper(n): # non-recursive
    if(n<=0):
        return 0;
    else:
        sum=0
        for i in range(n+1):
            sum+=i
        return sum

def helper_rec(n): # recursive
    if n<=0:
        return 0;
    else:
        return n+helper_rec(n-1)

n=int(input("Enter any number:\n"))
print(helper(n))
print(helper_rec(n))
```
